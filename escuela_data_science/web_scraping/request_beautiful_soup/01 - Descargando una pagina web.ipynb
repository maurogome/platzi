```python
article_list = [section.a.get('href') for section in s_seccion.find('ul', attrs={'class':'article-list'}).find_all('h2')]
article_list
```

```python
url_nota = article_list[0]
```

```python
url_nota
```

```python
try:
    nota = requests.get(url_nota)
    if nota.status_code == 200:
        s_nota = BeautifulSoup(nota.text, 'lxml')
        # Extraemos el titulo:
        titulo = s_nota.find('h1', attrs = {'class':'article-title'})
        print(titulo.text)
        # Extraemos la fecha:
        fecha = s_nota.find('span', attrs = {'pubdate':'pubdate'}).get('datetime')
        print(fecha)
        # Reto: Extraer la volanta, el copete, el cuerpo y el autor
        # Extraemos la volanta
        volanta = s_nota.find('h2', attrs = {'class' : 'article-prefix'})
        print(volanta.text)
        
        cuerpo = s_nota.find('div', attrs = {'class' : 'article-text'})
        print(cuerpo.text)
        
        
        
except Exception as e:
    print('Error:')
    print(e, end = '\n')
```

```python
media = s_nota.find('div', attrs = {'class':'article-main-media-image'})
```

```python
imagenes = media.find_all('img')
imagenes
```

```python
if len(imagenes) == 0:
    print('No se encotraron imagenes')
else:
    imagen = imagenes[-1]
    img_src = imagen.get('data-src')
    print(img_src)
```

```python
img_req = requests.get(img_src)
```
