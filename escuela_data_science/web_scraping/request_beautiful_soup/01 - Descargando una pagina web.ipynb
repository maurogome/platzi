```python
from bs4 import BeautifulSoup

# BeautifulSoup nos permite parsear el codigo html. Es decir identificar las partes del codigo que nos interesa
# El parser es quien separa el texto en pequeños fragmentos mas manejables
```

```python
s = BeautifulSoup(p12.text, 'lxml')
```

```python
type(s)
```

```python
print(s.prettify())
```
